```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import string
import warnings
warnings.simplefilter("ignore")
from statsmodels.formula.api import ols
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
import scipy.stats as stats
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_selection import RFE
```

```python
df = pd.read_csv("master_df.csv")
```

```python
df.head()
```

```python

df_features = df[['floors','age_when_sold','view']]
```

```python
df_features['bedrooms+bathrooms'] = df['bedrooms'] + df['bathrooms']
```

```python
df_features
```

```python
df_features_corr = df_features.corr()
df_features_corr
```

```python
abs(df_features_corr) > 0.50
```

```python
# Begin building the OLS model
```

```python
y = df['price']

x = df_features
```

```python
X = sm.add_constant(x)
```

```python
linreg = sm.OLS(y, X).fit()
print(linreg.summary())
```

```python
#These are my training and test sets of data.
X_train, X_test, y_train, y_test = train_test_split(df_features,df['price'],test_size=0.2)
```

```python
selector = RFE(linreg, n_features_to_select=1)
selector = selector.fit(X_train,y_train)
```
