```python
def remove_outliers(df, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound  = q1 - (2 * iqr)
    upper_bound = q3 + (2 * iqr)
    out_df = df.loc[(df[col] > lower_bound) & (df[col] < upper_bound)]
    return out_df
```

```python
rem_out_df = remove_outliers(df, 'price')
```

```python
rem_out_df = rem_out_df.drop(columns=['waterfront','sqft_lot','condition','yr_built','zipcode','sqft_lot15','yr_renovated','zipcode','age_when_sold','date','basement','year','Renovated','floors','view'])
```

```python
for col in rem_out_df.columns:
    remove_outliers(df, col)
```
